# Functions

```python
def my_func():
    print("Hello from my_func!")
```

```python
my_func()
```

```python
def my_func_p(param):
    print(f"Hello from my_func_p with param: {param}")
```

```python
my_func_p("This is a parameter")
```

```python
def my_func_p_default(param="default_value"):
    print(f"Hello from my_func_p_default with param: {param}")
```

```python
my_func_p_default
```

```python
def my_func_with_docstring():
    """This is a docstring for my_func_with_docstring."""
    print("Hello from my_func_with_docstring!")
```

```python
## Lambda Expressions
add = lambda x, y: x + y
result = add(5, 3)
print(result)
```

```python
def time_two(x):
    return x * 2

result = time_two(5)
print(result)
```
