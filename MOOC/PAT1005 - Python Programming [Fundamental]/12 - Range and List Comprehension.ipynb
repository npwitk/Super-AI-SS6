# `range()`

```python
range(0,5)
```

```python
for item in range(0,5):
    print(item)
```

```python
for i in range(0,200,10): # start at 0, stop before 200, step by 10
    print(i)
```

```python
list(range(20))
```

```python
list(range(11))
```

## List Comprehension

```python
# Original For Loop

x = [1,2,3,4]
out = []
for num in x:
    out.append(num**2)
print(out)
```

```python
# List Comprehension

[num ** 2 for num in x]
```
