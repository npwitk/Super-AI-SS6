# Built-in Python Expression
- map and filter

## Map

```python
seq = [1,2,3,4,5]

def time_two(var):
    return var * 2

lambda x: x * 2
```

```python
list(map(time_two, seq))
```

```python
list(map(lambda x: x * 2, seq))
```

## Filter

```python
seq = [1,2,3,4,5]

def is_even(num):
    return num % 2 == 0

list(filter(is_even, seq))
```

```python
list(filter(lambda x: x % 2 == 0, seq))
```
