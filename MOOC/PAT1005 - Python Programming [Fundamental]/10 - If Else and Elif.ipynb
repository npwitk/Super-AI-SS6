# Control Flow Statements

```python
if True:
    print("Hello")
```

```python
if False:
# print("Hello") ERROR: IndentationError: expected an indented block
    pass # Do nothing, just a placeholder for the code block
```

```python
if (1==2):
    print("Hello")
else:
    print("If statement is False, execute this block")
```

```python
if (1==2):
    print("Hello")
elif (2==4):
    print("If statement is False, execute this block")
else:
    print("All previous conditions are False, execute this block")
```
